```python
class Persona:

    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        print(f"Hola, mi nombre es {self.nombre} y tengo {self.edad} años.")


# Crear objetos
persona1 = Persona("Diego", 24)
persona2 = Persona("Ana", 21)

# Usar métodos
persona1.saludar()
persona2.saludar()
```

```python
# Encapsulación

class Cuenta:

    def __init__(self, saldo: float):
        self.__saldo = saldo

    def depositar(self, monto: float):
        self.__saldo += monto

    def retirar(self, monto: float):
        if monto > self.__saldo:
            return "Error, saldo insuficiente"

        self.__saldo -= monto

    def mostrar_saldo(self):
        return self.__saldo


cuenta = Cuenta(1000)

print(cuenta.mostrar_saldo())
```

```python
class Rectangulo:

    def __init__(self, base: float, altura: float) -> None:
        self.base = base
        self.altura = altura

    def area(self) -> float:
        return self.base * self.altura

    def perimetro(self) -> float:
        return 2 * (self.base + self.altura)

    def __str__(self) -> str:
        return f"Rectangulo de base: {self.base} y altura: {self.altura}"


mi_rectangulo = Rectangulo(3, 5)

print(mi_rectangulo.area())
print(mi_rectangulo.perimetro())
print(mi_rectangulo)
```

```python
# Ejercicio 5

class Libro:
    def __init__(self, titulo: str, autor: str, precio: float) -> None:
        self.titulo = titulo
        self.autor = autor
        self.__precio = precio

    def get_precio(self) -> float:
        return self.__precio

    def set_precio(self, nuevo_precio: float) -> None:
        if nuevo_precio > 0:
            self.__precio = nuevo_precio
        else:
            print(f"Error: el nuevo precio es 0 o negativo: {nuevo_precio}")


libro = Libro(
    titulo="Mi planta de naranja Lima",
    autor="Jose Mauro de Vasconcelos",
    precio=20
)

# print(libro.__precio)

print(libro.get_precio())

libro.set_precio(-50)

print(libro.get_precio())
```

```python
#Ejercicio 6
class Animal:
    def __init__(self, nombre: str, edad: int) -> None:
        self.nombre = nombre
        self.edad = edad

    def hacer_sonido(self):
        print("Grrrrrr")


class Perro(Animal):
    def __init__(self, nombre: str, edad: int) -> None:
        super().__init__(nombre, edad)

    def hacer_sonido(self) -> str:
        print("Gua gua!")


class Gato(Animal):
    def __init__(self, nombre: str, edad: int) -> None:
        super().__init__(nombre, edad)

    def hacer_sonido(self):
        print("Miau Miau!")


perro_1 = Perro("Hachiko", 3)
perro_2 = Perro("Fido", 5)
gato_1 = Gato("Lucifer", 10)
vaca_1 = Animal("Lola", 14)

animales = [perro_1, perro_2, gato_1, vaca_1]

for animal in animales:
    animal.hacer_sonido()
```

```python
# Ejercicio 8
class Reproductor:
    def __init__(self):
        self.playlist = []
        self.posicion = 0

    def agregar_cancion(self, cancion: str) -> None:
        self.playlist.append(cancion)

    def cancion_actual(self) -> str:
        if not self.playlist:
            return "No hay canciones en la playlist"

        return self.playlist[self.posicion]

    def siguiente(self) -> str:
        self.posicion += 1
        if self.posicion < len(self.playlist):
            return self.cancion_actual()

        self.posicion -= 1
        return self.cancion_actual()

    def anterior(self) -> str:
        self.posicion -= 1
        if self.posicion >= 0:
            return self.cancion_actual()

        self.posicion += 1
        return self.cancion_actual()


mi_reproductor = Reproductor()
mi_reproductor.agregar_cancion("The trooper")
mi_reproductor.agregar_cancion("Dance of death")
mi_reproductor.cancion_actual()
mi_reproductor.siguiente()
mi_reproductor.cancion_actual()
```

```python
# Ejercicio 9
class Motor:
    def __init__(self, tipo: str, caballos: float) -> None:
        self.tipo = tipo
        self.caballos = caballos
        self.encendido = False

    def arrancar(self) -> str:
        if self.encendido:
            return "El motor ya esta encendido"

        self.encendido = True
        return f"El motor {self.tipo} de {self.caballos} HP se encendio"


class Vehiculo:
    def __init__(self, marca: str, motor: Motor) -> None:
        self.marca = marca
        self.motor = motor

    def arrancar(self) -> str:
        return self.motor.arrancar()


motor = Motor("gasolina", 150)
vehiculo = Vehiculo("Ford", motor)

print(vehiculo.arrancar())
print(vehiculo.arrancar())
```

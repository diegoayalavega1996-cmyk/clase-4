```python
#Ejercicio 2

numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]
numeros_pares = []
numeros_impares = []
for numero in numeros:
    if numero % 2 == 0:
       numeros_pares.append(numero)

    if numero % 2 == 1:
        numeros_impares.append(numero)

print(f"numeros pares: {numeros_pares}")
print(f"numeros impares: {numeros_impares}")
    
```

```python
#ejercicio 4: Dada una lista de palabras, cuenta cuántas palabras tienen más de 5 letras

palabras = ['python', 'sol', 'programacion', 'lista', 'computadora', 'if', 'variable']
contador = 0
for palabra in palabras:
    if len(palabra) > 5:
        contador += 1

print(f"la cantidad de palabras con mas de 5 letras es: {contador}")
```

```python
# Ejercicio 5: Dada una lista de notas, calcula el promedio e indica si el estudiante aprobó o desaprobó.
notas = [14, 18, 11, 16, 9, 13, 20]
suma = 0
for nota in notas:
    suma = suma + nota
promedio = suma / len(notas)

print("Promedio:", promedio)

if promedio >= 11:
    print("El estudiante aprobó.")
else:
    print("El estudiante desaprobó.")
```

```python
#Ejercicio 9 Dadas dos listas, crea una nueva lista con los elementos que no se repiten entre ambas listas.
lista1 = [4, 8, 12, 16, 20] 
lista2 = [8, 10, 12, 18, 20]

no_comunes = []

for numero in lista1:
    if numero not in lista2:
        no_comunes.append(numero)


for numero2 in lista2:
    if numero2 not in lista1:
        no_comunes.append(numero2)

print(no_comunes)
```

```python
#Ejercicio 10 Invertir una lista sin usar reverse()

colores = ['rojo', 'azul', 'verde', 'amarillo', 'morado']

invertida = []

for i in range(len(colores)-1, -1, -1):
    invertida.append(colores[i])

print(invertida)
```

```python
#ejercicio 13 Dada una lista de nombres repetidos, muestra cuántas veces aparece cada nombre.
nombres = ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis', 'Ana', 'María', 'Carlos']

contados = []

for nombre in nombres:
    if nombre not in contados:
        cantidad = 0

        for n in nombres:
            if nombre == n:
                cantidad += 1

        print(nombre, "aparece", cantidad, "veces")
        contados.append(nombre)
```

```python
#ejercicio 16 Dada una lista de números, divídela en dos mitades.
numeros = [10, 20, 30, 40, 50, 60, 70, 80]
half_list = []
last_list = []

longitud = len(numeros)
indice_mitad = longitud / 2

for idx, numero in enumerate(numeros):
    if idx < indice_mitad:
        half_list.append(numero)
    else:
        last_list.append(numero)
print(f"priemra mitad: {half_list}")
print(f"ultima mitad: {last_list}")
```

```python
#Ejercicio 17 Dada una lista de palabras, reemplaza todas las apariciones de 'java' por 'python'.
palabras = ['java', 'python', 'java', 'c++', 'java', 'javascript']

nueva_lista = []

for palabra in palabras:
    nueva_palabra = palabra.replace("java", "python")
    nueva_lista.append(nueva_palabra)

print(f"Nueva lista: {nueva_lista}")
```

```python
#ejercicio 19 Dadas dos listas del mismo tamaño, crea una nueva lista alternando los elementos de ambas listas.

numeros = [1, 2, 3, 4]
letras = ['a', 'b', 'c', 'd']

resultado_esperado = [1, 'a', 2, 'b', 3, 'c', 4, 'd']

lista_alternada = []

for numero, letra in zip(numeros, letras):
    lista_alternada.extend([numero, letra])

print(f"Lista alternada: {lista_alternada}")
```

```python
#Ejercicio 20 Dada una lista de estudiantes y otra lista con sus notas, crea una nueva lista solo con los nombres de los estudiantes que aprobaron. Considera que la nota mínima aprobatoria es 11.
estudiantes = ['Ana', 'Luis', 'Carlos', 'María', 'Pedro']
notas = [18, 10, 14, 8, 16]

lista_aprobados = []

for estudiante, nota in zip(estudiantes, notas):
    if nota > 11:
        lista_aprobados.append(estudiante)

print(f"lista de aprobados: {lista_aprobados}")
```

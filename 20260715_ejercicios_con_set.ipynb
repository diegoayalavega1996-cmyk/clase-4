```python
#ejercicio1
estudiantes_python = {
    "Ana", "Luis", "María", "Carlos", "Sofía", "Diego"
}

estudiantes_javascript = {
    "María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"
}

# 1. Estudiantes inscritos en ambos cursos
ambos = estudiantes_python & estudiantes_javascript

# 2. Estudiantes que solo están en Python
solo_python = estudiantes_python - estudiantes_javascript

# 3. Estudiantes que solo están en JavaScript
solo_javascript = estudiantes_javascript - estudiantes_python

# 4. Todos los estudiantes sin repetir nombres
todos = estudiantes_python | estudiantes_javascript

# Mostrar resultados
print("Estudiantes en ambos cursos:", ambos)
print("Solo en Python:", solo_python)
print("Solo en JavaScript:", solo_javascript)
print("Todos los estudiantes:", todos)
```

```python
#ejercicio 2
almacen_arequipa = {
    "laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"
}

almacen_lima = {
    "laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"
}

# Productos disponibles en ambos almacenes
ambos = almacen_arequipa & almacen_lima

# Productos exclusivos del almacén de Arequipa
solo_arequipa = almacen_arequipa - almacen_lima

# Productos exclusivos del almacén de Lima
solo_lima = almacen_lima - almacen_arequipa

# Todos los productos sin repetir
todos = almacen_arequipa | almacen_lima

# Verificar si ambos almacenes tienen el mismo inventario
mismo_inventario = almacen_arequipa == almacen_lima

print("Productos en ambos almacenes:", ambos)
print("Productos exclusivos de Arequipa:", solo_arequipa)
print("Productos exclusivos de Lima:", solo_lima)
print("Todos los productos:", todos)
print("¿Tienen el mismo inventario?", mismo_inventario)
```

```python
# Ejercicio 3

permisos_usuario = {
    "leer", "editar", "comentar", "descargar"
}

permisos_requeridos = {
    "leer", "editar", "eliminar", "publicar"
}

# Verificar si el usuario tiene todos los permisos requeridos
tiene_todos = permisos_requeridos <= permisos_usuario

# Permisos que le faltan
faltan = permisos_requeridos - permisos_usuario

# Permisos adicionales que tiene
adicionales = permisos_usuario - permisos_requeridos

# Cantidad total de permisos diferentes
total_permisos = len(permisos_usuario | permisos_requeridos)

# Mostrar resultados
print("¿Tiene todos los permisos requeridos?:", tiene_todos)
print("Permisos que le faltan:", faltan)
print("Permisos adicionales:", adicionales)
print("Cantidad total de permisos diferentes:", total_permisos)
```

```python
#ejercicio 4

numeros_a = {2, 4, 6, 8, 10, 12, 14}
numeros_b = {4, 6, 9, 10, 13, 14, 16}
numeros_c = {1, 4, 6, 10, 11, 14, 18}

# Números presentes en los tres conjuntos
tres_conjuntos = numeros_a & numeros_b & numeros_c

# Números presentes en A y B, pero no en C
a_b_no_c = (numeros_a & numeros_b) - numeros_c

# Números que solo pertenecen a A
solo_a = numeros_a - numeros_b - numeros_c

# Todos los números diferentes
todos = numeros_a | numeros_b | numeros_c

# Mostrar resultados
print("Números presentes en los tres conjuntos:", tres_conjuntos)
print("Números presentes en A y B, pero no en C:", a_b_no_c)
print("Números que solo pertenecen a A:", solo_a)
print("Todos los números diferentes:", todos)
```

```python
# ejercicio 5

habilidades_candidato_1 = {
    "Python", "Git", "SQL", "Docker", "React"
}

habilidades_candidato_2 = {
    "Python", "JavaScript", "React", "Node.js", "Git"
}

habilidades_requeridas = {
    "Python", "Git", "SQL", "Docker"
}

# Habilidades compartidas
compartidas = habilidades_candidato_1 & habilidades_candidato_2

# Habilidades exclusivas de cada candidato
solo_candidato_1 = habilidades_candidato_1 - habilidades_candidato_2
solo_candidato_2 = habilidades_candidato_2 - habilidades_candidato_1

# Habilidades requeridas que le faltan a cada candidato
faltan_candidato_1 = habilidades_requeridas - habilidades_candidato_1
faltan_candidato_2 = habilidades_requeridas - habilidades_candidato_2

# Qué candidato cumple con más requisitos
cumple_1 = len(habilidades_candidato_1 & habilidades_requeridas)
cumple_2 = len(habilidades_candidato_2 & habilidades_requeridas)

if cumple_1 > cumple_2:
    mejor = "Candidato 1"
elif cumple_2 > cumple_1:
    mejor = "Candidato 2"
else:
    mejor = "Ambos cumplen la misma cantidad de requisitos"

# Todas las habilidades sin repetir
todas = habilidades_candidato_1 | habilidades_candidato_2

# Mostrar resultados
print("Habilidades compartidas:", compartidas)
print("Habilidades exclusivas del candidato 1:", solo_candidato_1)
print("Habilidades exclusivas del candidato 2:", solo_candidato_2)
print("Habilidades que le faltan al candidato 1:", faltan_candidato_1)
print("Habilidades que le faltan al candidato 2:", faltan_candidato_2)
print("Candidato que cumple con más requisitos:", mejor)
print("Todas las habilidades disponibles:", todas)
```

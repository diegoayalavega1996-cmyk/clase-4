```python
#TUPLAS
#Ejercicio 1
numeros = (15, 8, 42, 3, 27, 19, 5)

mayor = max(numeros)
menor = min(numeros)
diferencia = mayor - menor

print("Número mayor:", mayor)
print("Número menor:", menor)
print("Diferencia:", diferencia)
```

```python
#Ejercicio 2
datos = (5, 8, 5, 1, 9, 5, 2, 8, 5, 3)

numero = int(input("Ingrese un número: "))

cantidad = datos.count(numero)

print("El número", numero, "aparece", cantidad, "veces.")
```

```python
#Ejercicio 3
nombres = ("Ana", "Luis", "Pedro", "María", "Luis", "Carlos")

nombre_buscar = input("Ingrese un nombre: ")

if nombre_buscar in nombres:
    posicion = nombres.index(nombre_buscar)
    print("La primera posición es:", posicion)
else:
    print("El nombre no existe en la tupla.")
```

```python
#Ejercicio 4
notas = (18, 15, 20, 11, 14, 17, 19)

promedio = sum(notas) / len(notas)
nota_maxima = max(notas)
nota_minima = min(notas)

aprobados = 0

for nota in notas:
    if nota >= 11:
        aprobados += 1

print("Promedio:", promedio)
print("Nota máxima:", nota_maxima)
print("Nota mínima:", nota_minima)
print("Cantidad de aprobados:", aprobados)
```

```python
#Ejercicio 5
ventas = [
    ("Lunes", 120),
    ("Martes", 150),
    ("Miércoles", 95),
    ("Jueves", 180),
    ("Viernes", 210)
]

total = 0
mayor_venta = ventas[0]

for dia, venta in ventas:
    total += venta

    if venta > mayor_venta[1]:
        mayor_venta = (dia, venta)

promedio = total / len(ventas)

print("Día con mayor venta:", mayor_venta[0])
print("Mayor venta:", mayor_venta[1])
print("Total de ventas:", total)
print("Promedio de ventas:", promedio)
```

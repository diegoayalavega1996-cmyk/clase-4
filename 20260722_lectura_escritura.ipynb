```python
with open("./data/notas.txt", "w") as f:
    f.write("primera linea\n")
    f.write("segunda linea\n")


with open("./data/notas.txt", "a") as f:
    f.write("tercera linea\n")
```

```python
#lectura de archivos
with open("./data/notas.txt", "r") as f:
    contenido = f.read()
    print(contenido)
```

```python
with open("./data/notas.txt", "r") as f:
    contenido = f.readlines()
    print(contenido)
```

```python
with open("./data/notas.txt", "r") as f:
    contenido = f.readlines()
    for linea in contenido:
        print(linea)
```

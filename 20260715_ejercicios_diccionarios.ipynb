```python
#Ejercicio 1

inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3} 
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5} 
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}
```

```python

inventario_total = dict(inventario_norte)
# Sumar inventario_sur
for llave, valor in inventario_sur.items():
    print(f"Llave: {llave} valor: {valor}")
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor
print(inventario_total)
```

```python
# Sumar inventario_centro
for llave, valor in inventario_centro.items():
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor
#sorted(inventario_total)
print(inventario_total)
```

```python
# Contar en cuántas sucursales aparece cada producto
inventario_final = {}

for producto in sorted(inventario_total):
    sucursales = 0

    if producto in inventario_norte:
        sucursales += 1
    if producto in inventario_sur:
        sucursales += 1
    if producto in inventario_centro:
        sucursales += 1

    inventario_final[producto] = {
        "cantidad_total": inventario_total[producto],
        "sucursales": sucursales
    }

print(inventario_final)
```

```python
# Ejercicio 3

usuarios = {
    " ANA_01 ": {
        "nombre": " ana torres ",
        "correo": " ANA@MAIL.COM ",
        "edad": "24"
    },
    "LUIS_02": {
        "nombre": "luis perez",
        "correo": "Luis@mail.com",
        "edad": "treinta"
    },
    "carla_03": {
        "nombre": "CARLA RUIZ",
        "correo": "carla@mail.com",
        "edad": "27"
    }
}

usuarios_limpios = {}

for usuario, datos in usuarios.items():
    print(usuario, datos)

    llave_limpia = usuario.strip()
    correo_minuscula = datos["correo"].lower()
    nombre_limpio = datos["nombre"].strip().upper()

    try:
        edad_entera = int(datos["edad"])
    except:
        edad_entera = None

    datos_limpios = {
        "nombre": nombre_limpio,
        "correo": correo_minuscula,
        "edad": edad_entera
    }

    usuarios_limpios[llave_limpia] = datos_limpios

    print("=" * 20)

print(usuarios_limpios)
```
